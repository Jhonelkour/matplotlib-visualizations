```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
%matplotlib inline
```

```python
df = pd.read_csv('./Data/500_Person_Gender_Height_Weight_Index.csv')
```

```python
df.head()
```

```python
# Count the frequency of each gender in the 'Gender' column of the DataFrame
gender_freq = df['Gender'].value_counts()
gender_freq
```

```python
# Get the index (unique gender labels) from the gender frequency counts
gender_freq.index
```

```python
# Create a bar chart to visualize the frequency of each gender
# The x-axis represents the gender categories, and the y-axis represents the frequency of each gender
# The colors 'pink' and 'blue' are used for the bars to differentiate the genders visually
plt.bar(gender_freq.index, gender_freq, color=['pink', 'blue'])

# Set the title of the bar chart
plt.title('Gender Frequency')

# Set the labels for the x-axis and y-axis to describe the data
plt.xlabel('Gender')  # Label for the x-axis (gender categories)
plt.ylabel('Frequency')  # Label for the y-axis (frequency count of each gender)

# Display the plot on the screen
plt.show()
```

```python
# Define the data for the pie chart (Gender Frequency)
data = gender_freq

# Labels for each slice of the pie (Unique gender categories)
labels = gender_freq.index

# Define custom colors for each slice of the pie (pink for Female, blue for Male)
color = ['#FF69B4', '#1E90FF']

# Define how far each slice should "explode" or stand out from the center
# Explode the first slice to highlight it
explode = [0.1, 0]

# Create the pie chart
plt.pie(
    data,                             # Data representing the frequency of each gender
    labels=labels,                    # Add labels to each slice
    colors=color,                     # Set custom colors for each slice
    autopct='%1.1f%%',                # Display the percentage values for each slice
    explode=explode,                  # Offset the first slice from the center
    shadow=True,                      # Add a shadow effect to the pie chart
    startangle=80,                    # Rotate the start angle of the pie chart
    wedgeprops={'edgecolor': 'black'}  # Add a black edge around each slice
)

# Add a title to the pie chart
plt.title('Developer by Language')

# Display the pie chart with a legend
plt.legend()

# Show the plot
plt.show()
```
