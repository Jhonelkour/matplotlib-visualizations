```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
%matplotlib inline
```

```python
ages = [10, 30, 45, 30, 31, 46, 12, 16, 32, 23, 28, 29]
bins = [10, 20, 30, 40, 50, 60]

mean = np.mean(ages)
print(mean)
plt.hist(ages, edgecolor='black', bins=bins)
plt.axvline(mean, color='red', linestyle='--', linewidth=3, label='Mean')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
