```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Define the data for the pie chart (Developers By Language)
data = [70,20,10,50]

# Labels for each slice of the pie
labels = ['Python', 'Javascript', 'HTML', 'Java']

# Define custom colors for each slice
color = ['#008fd5','#fc4f30','#e5ae37','#6d904f']

# Set how far each slice should "explode" or stand out from the center
explode = [0.1,0,0.1,0]

plt.pie(
        data,                           
        labels=labels,                      # Add labels to each slice
        colors=color,                       # Set custom colors for each slice
        autopct='%1.1f%%',                  # Display percentage values for each slice
        explode=explode,                    #  Offset selected slices from the center
        shadow= TRUE,                       # Add a shadow effect to the pie chart
        startangle=80,                      # Rotate the start angle of the pie chart
        wedgeprops={'edgecolor':'black'}    # Add a black edge around each slice
        )

# Add a title to the pie chart
plt.title('Developer by Language')
# Display the pie chart
plt.show()
```
